```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

trainset = pd.read_csv('Google_Stock_Price_Train.csv')
trainset = trainset.iloc[:,1:2].values  # open stock price
testset = pd.read_csv('Google_Stock_Price_Test.csv')
testset_input = testset.iloc[:,1:2].values  # open stock price

# trainset

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
trainset = scaler.fit_transform(trainset)
testset_input = scaler.transform(testset_input)

x_train = trainset[0:1257]
y_train = trainset[1:1258]

x_train = np.reshape(x_train,(1257,1,1))
testset_input = np.reshape(testset_input,(20,1,1))
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM

# initializing the RNN

regressor = Sequential()
regressor.add(LSTM(units=4,activation='sigmoid',input_shape=(None,1)))
regressor.add(Dense(units=1))
regressor.compile(optimizer='adam',loss='mean_squared_error')
regressor.fit(x_train,y_train,batch_size=32,epochs=200)
```

```python
# predicting and visualizing
predicted = regressor.predict(testset_input)
predicted = scaler.inverse_transform(predicted)

plt.plot(predicted)
plt.plot(testset.iloc[:,1:2].values,"--b")
plt.plot()
```

```python
import math
from sklearn.metrics import mean_squared_error
rmse = math.sqrt(mean_squared_error(testset.iloc[:,1:2].values,predicted))
rmse/800
```
