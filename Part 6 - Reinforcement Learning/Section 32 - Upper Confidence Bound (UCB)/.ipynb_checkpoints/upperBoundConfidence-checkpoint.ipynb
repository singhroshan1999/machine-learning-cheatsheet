```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
dataset
```

```python
# implementing UCB from scratch
import math
N = 10000
d = 10
ads_selected = []
number_of_selection = [0] * d
sum_of_reward = [0] * d
total_reward = 0
for n in range(0,N):
    max_upper_bound = 0
    ads = 0
    for i in range(0,d):
        if number_of_selection[i] > 0:
            average_reward = sum_of_reward[i]/number_of_selection[i]
            delta_i = math.sqrt(3/2*math.log(n+1)/number_of_selection[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound = 1e400
        if max_upper_bound < upper_bound:
            max_upper_bound = upper_bound
            ads = i
    ads_selected.append(ads)
    number_of_selection[ads] += 1
    reward = dataset.values[n, ads]
    sum_of_reward[ads] += reward
    total_reward += reward 
# print(ads_selected)
print(number_of_selection)
print(sum_of_reward)
print(total_reward)
```

```python
# visualizing Results
plt.hist(ads_selected)
plt.title('Histogram of Selected ads')
plt.xlabel('Ads')
plt.ylabel('Number of times Selected')
plt.show()
```
