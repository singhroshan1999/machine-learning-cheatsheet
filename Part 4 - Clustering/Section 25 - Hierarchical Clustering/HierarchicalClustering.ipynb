```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,3:5].values
# dataset
```

```python
# using Dendrogram to find number of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x ,method = 'ward'))
# linkage is algo of hierarchial clustering
#method = 'ward' minimises variance in each clusters
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# creating Hierarchical Clusters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5,affinity = 'euclidean',linkage = 'ward')
y_predicted_clusters = hc.fit_predict(x)
pd.Series(y_predicted_clusters)
```

```python
# Visualizing predicted clusters
plt.plot(x[y_predicted_clusters == 0, 0],x[y_predicted_clusters == 0, 1],'or',label = '1')
plt.plot(x[y_predicted_clusters == 1, 0],x[y_predicted_clusters == 1, 1],'og',label = '2')
plt.plot(x[y_predicted_clusters == 2, 0],x[y_predicted_clusters == 2, 1],'ob',label = '3')
plt.plot(x[y_predicted_clusters == 3, 0],x[y_predicted_clusters == 3, 1],'om',label = '4')
plt.plot(x[y_predicted_clusters == 4, 0],x[y_predicted_clusters == 4, 1],'oc',label = '5')
plt.ylabel('Spending Score')
plt.xlabel('Annual Income')
plt.title('Clusters of clients')
# plt.legend()
plt.show()
```
