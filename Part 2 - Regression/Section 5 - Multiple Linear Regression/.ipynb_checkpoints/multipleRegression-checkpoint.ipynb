```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split

#prepairing data
dataset = pd.read_csv('50_Startups.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values


from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
x[:,3] = labelencoder.fit_transform(x[:,3]) 
onehotencoder = OneHotEncoder(categorical_features = [3])
x = onehotencoder.fit_transform(x).toarray()
x = x[:,1:]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#linear regression modal
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#predicting test set
y_predict = regressor.predict(x_test)
print(y_test)
print(y_predict)
```

```python
#comparing y_test and Y_predict
xt = np.arange(1,y_test.size+1)
plt.plot(xt,y_test,"--g",label="y_test")
plt.plot(xt,y_predict,label="y_predict")
plt.title("test Vs Predict")
plt.legend()
plt.show()
```

## Backward elmenation

```python
#backward elmination
import statsmodels.formula.api as sm
x = np.append(arr = np.ones((50,1)).astype(int),values = x,axis = 1)
```

```python
# step1 fit full model
x_opt = x[:,[0,1,2,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = x_opt).fit()
# step2 getting P and comparing P and SL
regressor_OLS.summary()

x_opt = x[:,[0,1,3,4,5]]  # remove 2
regressor_OLS = sm.OLS(endog = y, exog = x_opt).fit()
regressor_OLS.summary()

x_opt = x[:,[0,3,4,5]]  # remove 1
regressor_OLS = sm.OLS(endog = y, exog = x_opt).fit()
regressor_OLS.summary()

x_opt = x[:,[0,3,5]]  # remove 4
regressor_OLS = sm.OLS(endog = y, exog = x_opt).fit()
regressor_OLS.summary()

x_opt = x[:,[0,3]]  # remove 5
regressor_OLS = sm.OLS(endog = y, exog = x_opt).fit()
regressor_OLS.summary()

# step FIN creating new regressor
regressor2 = LinearRegression()
regressor2.fit(x_train[:,[2]],y_train)
y_predict2 = regressor2.predict(x_test[:,[2]])
# comparing predicted vs test value
xt = np.arange(1,y_test.size+1)
plt.plot(xt,y_test,"--g",label="y_test")
plt.plot(xt,y_predict2,label="y_predict")
plt.title("test Vs Predict")
plt.legend()
plt.show()
```
